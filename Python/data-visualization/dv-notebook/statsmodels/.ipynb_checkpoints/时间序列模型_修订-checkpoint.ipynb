```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
# 划分训练集和测试集
df = pd.read_csv('./data/uswc.csv')
df.head()
```

```python
df_example=df[(df['warehouse_code']=='GBLTNA') & (df['order_date']>'2020-06-01')].sort_values(by='order_date')
```

```python
# 划分训练集和测试集
total_len = len(df_example)
test_len = 10
train, test = df_example[0:total_len - test_len-1], df_example[total_len - test_len-1:-1]
# 预测集长度等同测试集，用来评估模型准确度
forecast_len = test_len
```

```python
# 缺失值填充
df_example['Timestamp'] = pd.to_datetime(df_example['order_date'], format='%Y-%m-%d %H:%M:%S')
df_example.index = df_example['Timestamp']
df_example = df_example.resample('D').mean().fillna(0)

train['Timestamp'] = pd.to_datetime(train['order_date'], format='%Y-%m-%d %H:%M:%S')
train.index = train['Timestamp']
train = train.resample('D').mean().fillna(0)

test['Timestamp'] = pd.to_datetime(test['order_date'], format='%Y-%m-%d %H:%M:%S')
test.index = test['Timestamp']
test = test.resample('D').mean().fillna(0)
```

```python
# 模型训练
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['order_num']), seasonal_periods=7, trend='add', seasonal='add', ).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(forecast_len)
plt.figure(figsize=(16, 8))
plt.plot(train['order_num'], label='Train')
plt.plot(test['order_num'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

```python
y_hat_avg['diff']=y_hat_avg['Holt_Winter']/y_hat_avg['order_num']-1
```
